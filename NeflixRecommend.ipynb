```python
best_movie = pd.read_csv("./data/bestmovie.csv")
```

```python
best_show = pd.read_csv("./data/bestshow.csv")
```

```python
#CONDITIONS
try:
    watch = int(input("What would you like to watch\nEnter 1 for Shows\nEnter 2 for Movies:"))

    if watch == 1:
        data = best_show
        wh = "Shows"
    elif watch == 2:
        data = best_movie
        wh = "Movies"
    else:
        print("\nINVALID INPUT !!!!!!!!!!!")

    min_rating = float(input("Enter minimum rating : "))
    main_genre = input("Enter Genre : ")
    rel_year = int(input("After which year : "))

except ValueError:
    print("\nINVALID INPUT !!!!!!!!!!!")
```

```python
watch_list = []
no_of_movies = 5
for i in range(0,len(data)):
    scr = data.loc[i,"SCORE"]
    genre = data.loc[i,"MAIN_GENRE"]
    year = data.loc[i,"RELEASE_YEAR"]
    if len(watch_list) < no_of_movies:
        if scr >= min_rating and genre == main_genre and year >= rel_year:
            movie = data.loc[i,"TITLE"]
            watch_list.append(movie)
    elif len(watch_list) == 5:
        break
    
```

```python
num = 1
print(f"Here Are Some Recommended {wh} : ")
for i in watch_list:
    print(f"{num}. {i}")
    num += 1
```
